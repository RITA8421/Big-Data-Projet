## Part 1

```python
#Big Data
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as py

origin=pd.read_csv("/Users/suhao/Downloads/consommation-electrique-par-secteur-dactivite-commune.csv",delimiter=";")

#print(origin.columns)

consoParSec = DataFrame(origin,columns=['Année','Nom région','Nb sites Résidentiel',
                                            'Nb sites Professionnel','Nb sites Agriculture','Nb sites Industrie',
                                            'Nb sites Tertiaire','Nb sites Secteur non affecté'])

#Part 1 : Sites number of each region in France From 2011 to 2016
a1 = consoParSec[consoParSec['Nom région'].isin(['Auvergne-Rhône-Alpes'])]
a1 = a1.groupby(a1['Année']).sum()
a1.plot(marker='*')
for x,y in zip(a1.index, a1['Nb sites Résidentiel']):
    plt.text(x, y+10, '%.0f' % y, ha = 'center', va = 'bottom')
plt.grid()
plt.title('Number of sites in Auvergne-Rhône-Alpes from 2011 to 2016')  
plt.xlabel('year')  
plt.ylabel('number')

# from this chart, the sites numbers from 2011 to 2015 remained stable, 
# but during 2016, the number drops visibly, especially in the resident sites,
# from 3017691 to 2400704, we think this is not normal, it may caused by the loss of data
```

## Part 2
consomation of electricity in Île-de-France from 2011 to 2016 

```python
# part 2
# we take 'île de france ' as a example
consoInDiffArea = DataFrame(origin,columns=['Année','Nom région','Conso totale Résidentiel (MWh)',
        'Conso totale Professionnel (MWh)', 
        'Conso totale Agriculture (MWh)',
        'Conso totale Industrie (MWh)',
       'Conso totale Tertiaire (MWh)',
       'Conso totale Secteur non affecté (MWh)'])

b1 = consoInDiffArea[consoInDiffArea['Nom région'].isin(['Île-de-France'])]
b1 = b1.groupby(b1['Année']).sum()
b1

b1.plot(marker='*')
plt.title('Conso in different sector',verticalalignment='bottom')
plt.legend(loc='upper center', bbox_to_anchor=(0,1.5),ncol=3,fancybox=True,shadow=True)

plt.style.use('seaborn-whitegrid')

plt.xlabel('year')  
plt.ylabel('Conso totale(MWh)')
```

Conso in different sector in île de france from 2011 to 2016
in this chart, we can know that the highest consomation is "conso tertiaire", it remained stable from 2011 to 2016, during 2015-2016, the "conso industrie" grow higher than "conso professionnel", from 2012 to 2016, the "conso residentiel" drops continuously, it might caused by the the drop of residentiel site number.  

```python
from  matplotlib import cm
b2 = consoInDiffArea[consoInDiffArea['Nom région'].isin(['Île-de-France'])]
b2 = b2.groupby(b2['Année']).sum()
b2['non-résidentiel'] = b2['Conso totale Professionnel (MWh)']+b2['Conso totale Agriculture (MWh)']+b2['Conso totale Industrie (MWh)']+b2['Conso totale Tertiaire (MWh)']+b2['Conso totale Secteur non affecté (MWh)']
deleInfo = [1,2,3,4,5]
b2.drop(b2.columns[deleInfo], axis=1, inplace=True)


plt.subplots(figsize=(6,6))
plt.pie([b2.loc[2011][0],b2.loc[2011][1]], 
        explode = [0.1,0.1], labels=['Conso totale Résidentiel (MWh)','Conso totale non-résidentiel(MWh)'],  
        colors=['limegreen','powderblue'],  
        autopct='%1.2f%%', pctdistance=0.6, shadow=False,  
        labeldistance=1.1, startangle=None, radius=None,  
        counterclock=True, wedgeprops=None, textprops=None
       ) 
plt.title("Conso totale Résidentiel and Conso totale non-résidentiel(MWh) at île de France in 2011")
```

There are two pie chart that shows the proportion of conso totale résidentiel and non-résidentiel in 2011 and 2016.
